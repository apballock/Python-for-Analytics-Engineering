## Level 3 – Functions
### Exercise 1 – Basic Functions
Functions are reusable blocks of code defined with `def`.
They take inputs (parameters) and return outputs (return values).

```python
# Task 1 — Basic function with no parameters
def greet_analyst():
    """Prints a welcome message."""
    print("Welcome to the Financial Analysis Dashboard!")

greet_analyst()

# Task 2 — Function with parameters
def calculate_profit_margin(sales, profit):
    """
    Calculate profit margin percentage.
    
    Args:
        sales (float): Total sales amount
        profit (float): Total profit amount
    
    Returns:
        float: Profit margin as a percentage
    """
    if sales == 0:
        return 0
    return round((profit / sales) * 100, 2)

print(calculate_profit_margin(118726, 16893))
print(calculate_profit_margin(0, 1000))
```

```python
# Task 3 — Function with default parameter
def classify_performance(margin, threshold=15.0):
    """Classify performance as above or below threshold."""
    if margin >= threshold:
        return f"✅ Above target ({margin}% >= {threshold}%)"
    else:
        return f"❌ Below target ({margin}% < {threshold}%)"

print(classify_performance(18.5))
print(classify_performance(12.3))
print(classify_performance(12.3, threshold=10.0))

# Task 4 — Function returning multiple values
def sales_summary(sales_list):
    """Return min, max, and average of a sales list."""
    return min(sales_list), max(sales_list), round(sum(sales_list)/len(sales_list), 2)

sales = [6000, 2850, 3600, 1800, 2250]
minimum, maximum, average = sales_summary(sales)
print(f"Min: {minimum}, Max: {maximum}, Average: {average}")
```

## Exercise 2 – Functions with Data Structures
Real data analysis always combines functions with data structures.

```python
# Dataset
sales_data = [
    {"product": "Paseo", "segment": "Government", "country": "Spain", 
     "sales": 6000, "profit": 1200},
    {"product": "VTT", "segment": "Enterprise", "country": "France", 
     "sales": 2850, "profit": -285},
    {"product": "Velo", "segment": "Midmarket", "country": "Germany", 
     "sales": 3600, "profit": 540},
    {"product": "Amarilla", "segment": "Small Business", "country": "Canada", 
     "sales": 1800, "profit": 360},
    {"product": "Montana", "segment": "Channel Partners", "country": "USA", 
     "sales": 2250, "profit": 675},
]

# Task 1 — Total sales function
def total_sales(data):
    """Return total sales from a list of sales records."""
    return sum(record["sales"] for record in data)

print(f"Total Sales: €{total_sales(sales_data)}")

# Task 2 — Filter by segment
def filter_by_segment(data, segment):
    """Return records matching the given segment."""
    return [r for r in data if r["segment"] == segment]

government = filter_by_segment(sales_data, "Government")
print(f"Government records: {government}")
```

```python
# Task 3 — Find top performer
def top_performer(data, metric="sales"):
    """Return the record with the highest value for a given metric."""
    return max(data, key=lambda x: x[metric])

print(f"Top by sales: {top_performer(sales_data, 'sales')['product']}")
print(f"Top by profit: {top_performer(sales_data, 'profit')['product']}")

# Task 4 — Generate a summary report
def generate_report(data):
    """Print a formatted summary report of sales data."""
    print("=" * 50)
    print("SALES PERFORMANCE REPORT")
    print("=" * 50)
    for record in sorted(data, key=lambda x: x["sales"], reverse=True):
        margin = calculate_profit_margin(record["sales"], record["profit"])
        status = "✅" if record["profit"] > 0 else "❌"
        print(f"{status} {record['product']:10} | {record['country']:10} | "
              f"Sales: €{record['sales']:,} | Margin: {margin}%")
    print("=" * 50)
    print(f"TOTAL SALES: €{total_sales(data):,}")

generate_report(sales_data)
```

## Exercise 3 – Best Practices & Mini Project: Currency Converter
Best practices: single responsibility, type hints, docstrings, clear naming.

```python
# ❌ BAD function — does too many things, no docstring, unclear name
def f(x, y, z):
    r = x / y * z
    print(r)
    return r

# ✅ GOOD function — single responsibility, type hints, docstring
def convert_currency(amount: float, rate: float) -> float:
    """
    Convert an amount from one currency to another.
    
    Args:
        amount (float): The amount to convert
        rate (float): The exchange rate (target/source)
    
    Returns:
        float: The converted amount, rounded to 2 decimal places
    
    Raises:
        ValueError: If amount or rate is negative
    """
    if amount < 0 or rate <= 0:
        raise ValueError("Amount must be non-negative and rate must be positive")
    return round(amount * rate, 2)

# Task 1 — Test the converter
print(convert_currency(100, 1.08))   # EUR to USD
print(convert_currency(500, 0.85))   # USD to EUR
```

```python
# Task 2 — Full currency converter with exchange rates dictionary
exchange_rates = {
    "EUR_USD": 1.08,
    "EUR_GBP": 0.86,
    "EUR_BRL": 5.45,
    "USD_EUR": 0.93,
    "USD_GBP": 0.79,
    "GBP_EUR": 1.16,
}

def smart_converter(amount: float, from_currency: str, to_currency: str) -> str:
    """
    Convert between currencies using a rates dictionary.
    
    Args:
        amount: Amount to convert
        from_currency: Source currency code (e.g. 'EUR')
        to_currency: Target currency code (e.g. 'USD')
    
    Returns:
        str: Formatted conversion result
    """
    key = f"{from_currency}_{to_currency}"
    
    try:
        rate = exchange_rates[key]
        result = convert_currency(amount, rate)
        return f"{amount} {from_currency} = {result} {to_currency} (rate: {rate})"
    except KeyError:
        return f"❌ Exchange rate for {from_currency} → {to_currency} not available"
    except ValueError as e:
        return f"❌ Invalid input: {e}"

# Task 3 — Test the smart converter
print(smart_converter(1000, "EUR", "USD"))
print(smart_converter(1000, "EUR", "BRL"))
print(smart_converter(1000, "EUR", "JPY"))
print(smart_converter(-50, "EUR", "USD"))
```

```python
# Task 4 — Convert Financial Sample total sales to multiple currencies
total_eur = 118726.35
currencies = ["USD", "GBP", "BRL"]

print("\n--- Financial Sample Total Sales in Multiple Currencies ---")
for currency in currencies:
    print(smart_converter(total_eur, "EUR", currency))
```
