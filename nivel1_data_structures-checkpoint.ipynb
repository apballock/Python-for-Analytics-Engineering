## Level 1 – Data Structures
### Exercise 1 – Lists
Lists are ordered, mutable collections. They allow duplicate values.

```python
# Task 1 — Create a list of 5 products from the Financial Sample dataset
products = ["Paseo", "VTT", "Velo", "Amarilla", "Montana"]
print(products)
```

```python
# Task 2 — Add a new product to the end of the list
products.append("Carretera")
print(products)

# Task 3 — Remove the second product (index 1)
products.pop(1)
print(products)

# Task 4 — Sort the list alphabetically
products.sort()
print(products)
```

```python
# Task 5 — Print only the first 3 products using slicing
print(products[:3])

# Task 6 — Check if "Paseo" is in the list
print("Paseo" in products)

# Task 7 — Print the total number of products
print(f"Total products: {len(products)}")
```

## Exercise 2 – Tuples
Tuples are ordered but IMMUTABLE — once created, they cannot be changed.
Use tuples for data that should never be modified.

```python
# Task 1 — Create a tuple with country names from the dataset
countries = ("Spain", "France", "Germany", "Canada", "USA", "Mexico")
print(countries)

# Task 2 — Access the third country (index 2)
print(countries[2])

# Task 3 — Slice the first 3 countries
print(countries[:3])

# Task 4 — Check how many times "France" appears
print(countries.count("France"))
```

```python
# Task 5 — Find the index of "USA"
print(countries.index("USA"))
```

```python
# Task 6 — Try to modify a tuple (this will raise an error — that's expected!)
try:
    countries[0] = "Italy"
except TypeError as e:
    print(f"Error: {e}")
    print("Tuples are immutable — this is expected behavior!")

# Task 7 — Convert the tuple to a list, add "Italy", convert back to tuple
countries_list = list(countries)
countries_list.append("Italy")
countries = tuple(countries_list)
print(countries)
```

## Exercise 3 – Sets
Sets are UNORDERED collections with NO DUPLICATES.
Perfect for finding unique values, intersections, and differences.

```python
# Task 1 — Create a set of segments (with intentional duplicates)
segments_raw = ["Government", "Small Business", "Enterprise", 
                "Midmarket", "Government", "Small Business", "Channel Partners"]
segments = set(segments_raw)
print(f"Unique segments: {segments}")
print(f"Original: {len(segments_raw)} items → Unique: {len(segments)} items")

# Task 2 — Create two sets: profitable and high-volume segments
profitable = {"Channel Partners", "Government", "Midmarket"}
high_volume = {"Government", "Small Business", "Enterprise"}

# Task 3 — Intersection: BOTH profitable AND high-volume
both = profitable & high_volume
print(f"Both profitable and high-volume: {both}")

# Task 4 — Difference: profitable BUT NOT high-volume
profitable_only = profitable - high_volume
print(f"Profitable but not high-volume: {profitable_only}")

# Task 5 — Union: ALL unique segments
all_segments = profitable | high_volume
print(f"All segments combined: {all_segments}")

# Task 6 — Check if "Enterprise" is in the profitable set
print("Enterprise" in profitable)
```

## Exercise 4 – Dictionaries
Dictionaries store data as KEY:VALUE pairs.
Think of them like a mini-database row or a JSON object.

```python
# Task 1 — Create a dictionary representing one product
product = {
    "name": "Paseo",
    "category": "Bikes",
    "unit_price": 120.00,
    "units_sold": 50,
    "country": "Spain"
}
print(product)

# Task 2 — Access the product name and price
print(f"Product: {product['name']}, Price: €{product['unit_price']}")

# Task 3 — Add a new key: total_revenue
product["total_revenue"] = product["unit_price"] * product["units_sold"]
print(f"Total revenue: €{product['total_revenue']}")

# Task 4 — Update the units_sold
product["units_sold"] = 75
print(f"Updated units sold: {product['units_sold']}")
```

```python
# Task 5 — Create a list of dictionaries (like a database table)
sales_data = [
    {"product": "Paseo", "country": "Spain", "sales": 6000},
    {"product": "VTT", "country": "France", "sales": 2850},
    {"product": "Velo", "country": "Germany", "sales": 3600},
    {"product": "Amarilla", "country": "Canada", "sales": 1800},
    {"product": "Montana", "country": "USA", "sales": 2250}
]

# Task 6 — Print only products with sales > 3000
for item in sales_data:
    if item["sales"] > 3000:
        print(f"{item['product']} ({item['country']}): €{item['sales']}")

# Task 7 — Calculate total sales
total = sum(item["sales"] for item in sales_data)
print(f"Total sales: €{total}")
```
