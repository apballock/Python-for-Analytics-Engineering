## Level 2 – Control Flow
### Exercise 1 – Conditional Statements
if/elif/else lets us execute different code depending on conditions.

```python
# Task 1 — Classify profit margin
def classify_margin(margin):
    if margin > 20:
        return "Excellent"
    elif margin > 10:
        return "Good"
    elif margin > 0:
        return "Low"
    else:
        return "Loss — Negative Margin"

# Test with different values
test_margins = [25.3, 14.2, 5.8, -3.1]
for m in test_margins:
    print(f"Margin {m}% → {classify_margin(m)}")
```

```python
# Task 2 — Discount band classifier
def discount_band(discount_pct):
    if discount_pct == 0:
        return "None"
    elif discount_pct <= 10:
        return "Low"
    elif discount_pct <= 20:
        return "Medium"
    else:
        return "High"

print(discount_band(0))
print(discount_band(8))
print(discount_band(15))
print(discount_band(30))
```

## Exercise 2 – For Loops & List Comprehension
For loops iterate over any sequence. 
List comprehensions are a concise way to create lists from loops.

```python
# Task 1 — Loop through products and print each one
products = ["Paseo", "VTT", "Velo", "Amarilla", "Montana"]
for product in products:
    print(f"Processing: {product}")

# Task 2 — Loop with index using enumerate()
for i, product in enumerate(products):
    print(f"{i+1}. {product}")
```

```python
# Task 3 — Loop through a dictionary
sales = {"Paseo": 6000, "VTT": 2850, "Velo": 3600, "Amarilla": 1800}
for product, amount in sales.items():
    print(f"{product}: €{amount}")

# Task 4 — List comprehension: products with sales > 3000
high_sales = [p for p, s in sales.items() if s > 3000]
print(f"High sales products: {high_sales}")

# Task 5 — List comprehension: prices with 10% tax
prices = [120, 95, 45, 30, 150]
prices_with_tax = [round(p * 1.10, 2) for p in prices]
print(f"Prices with tax: {prices_with_tax}")
```

```python
# Task 6 — Nested loop: product x country combinations
top_products = ["Paseo", "VTT"]
top_countries = ["Spain", "France", "Germany"]
combinations = [(p, c) for p in top_products for c in top_countries]
for combo in combinations:
    print(combo)
```

## Exercise 3 – While Loops & Input Validation
While loops repeat as long as a condition is True.
They're perfect for input validation and retry logic.

```python
# Task 1 — Basic while loop: countdown
count = 5
while count > 0:
    print(f"Processing batch {count}...")
    count -= 1
print("Done!")

# Task 2 — Simulate input validation with a list of test inputs
test_inputs = ["abc", "-5", "150", "25.5"]
valid_margin = None
for test in test_inputs:
    try:
        value = float(test)
        if 0 <= value <= 100:
            valid_margin = value
            print(f"✅ Valid input: {value}")
            break
        else:
            print(f"❌ Out of range: {value}")
    except ValueError:
        print(f"❌ Not a number: {test}")

print(f"Final valid margin: {valid_margin}")
```

## Exercise 4 – Try / Except
try/except catches errors and lets us handle them gracefully
instead of crashing the program.

```python
# Task 1 — Basic try/except
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"

print(safe_divide(100, 4))
print(safe_divide(100, 0))

# Task 2 — Multiple except blocks
def parse_sales_data(value):
    try:
        return float(value)
    except ValueError:
        return f"Error: '{value}' is not a valid number"
    except TypeError:
        return "Error: Input cannot be None"

test_values = ["1500.50", "abc", None, "2000"]
for v in test_values:
    print(f"Input: {v} → Result: {parse_sales_data(v)}")
```

```python
# Task 3 — Try/except/else/finally
def load_dataset(filename):
    try:
        with open(filename, 'r') as f:
            data = f.read()
    except FileNotFoundError:
        print(f"❌ File '{filename}' not found")
        return None
    else:
        print(f"✅ File loaded successfully")
        return data
    finally:
        print("Load attempt complete")

load_dataset("financial_sample.csv")
load_dataset("nonexistent_file.csv")
```
